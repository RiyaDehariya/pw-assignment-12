```python
6. What are the most popular product keywords?
```

```python
import pandas as pd
import re
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
# Lowercase all product names and remove special characters
df['product_name_clean'] = df['product_name'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', str(x).lower()))
```

```python
# Inspect the cleaned product names
df['product_name_clean'].head()
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```python
# Set of stopwords to filter out
stop_words = set(stopwords.words('english'))
```

```python
# Tokenize and remove stopwords
def tokenize_and_remove_stopwords(text):
    tokens = word_tokenize(text)
    return [word for word in tokens if word not in stop_words]
```

```python
# Apply the function to the cleaned product names
df['tokens'] = df['product_name_clean'].apply(tokenize_and_remove_stopwords)
```

```python
# Inspect tokenized product names
df['tokens'].head()
```

```python
# Flatten the list of tokens into a single list
all_keywords = [keyword for sublist in df['tokens'] for keyword in sublist]
```

```python
# Count the frequency of each keyword
keyword_counts = Counter(all_keywords)
```

```python
# Convert the counts to a DataFrame for easy manipulation
keyword_df = pd.DataFrame(keyword_counts.items(), columns=['keyword', 'count'])
```

```python
# Sort the keywords by frequency
keyword_df = keyword_df.sort_values(by='count', ascending=False)
```

```python
# Display the top 10 most popular keywords
top_10_keywords = keyword_df.head(10)
print(top_10_keywords)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the top 10 most frequent keywords
plt.figure(figsize=(10, 6))
sns.barplot(x='count', y='keyword', data=top_10_keywords, palette='Blues_d')
plt.title('Top 10 Most Popular Product Keywords')
plt.xlabel('Keyword Frequency')
plt.ylabel('Keyword')
plt.show()
```

```python
7.For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist.
```

```python
# Group by artist and calculate average popularity
artist_popularity_avg = df.groupby('Artist')['Popularity'].mean().reset_index()

# Sort artists by average popularity in descending order (most popular first)
artist_popularity_avg = artist_popularity_avg.sort_values(by='Popularity', ascending=False)

# Get the top 5 most popular artists
top_5_artists = artist_popularity_avg.head(5)['Artist'].tolist()

# Display the top 5 artists
print("Top 5 Most Popular Artists:")
print(top_5_artists)
```

```python
# Create an empty list to store the most popular track for each artist
most_popular_tracks = []

# Loop through the top 5 artists and find their most popular track
for artist in top_5_artists:
    # Filter the dataset for the current artist
    artist_tracks = df[df['Artist'] == artist]

    # Find the track with the highest popularity for the current artist
    most_popular_track = artist_tracks.sort_values(by='Popularity', ascending=False).iloc[0]

    # Append the artist name and track name to the list
    most_popular_tracks.append({
        'Artist': artist,
        'Track Name': most_popular_track['Track Name'],
        'Popularity': most_popular_track['Popularity']
    })

# Convert the result to a DataFrame and display
most_popular_tracks_df = pd.DataFrame(most_popular_tracks)
print("Most Popular Tracks for the Top 5 Artists:")
print(most_popular_tracks_df)
```
